```python
import os
import PIL
import numpy as np
import pandas as pd

from PIL import Image, ImageFilter
from matplotlib import pyplot as plt
```

# Images exploration

```python
path = '/home/ksawi/Documents/Workspace/carla/carla_racetrack_BS/data/experiments/circut_spa/20200603_2058/MPCController_depth_rgb_collisions_2794/sensors'
imgs = os.listdir(path)
rgb = sorted([img for img in imgs if 'rgb' in img])
depth = sorted([img for img in imgs if 'depth' in img])
len(rgb), len(depth)
```

```python
rgb[100:104], depth[100:104]
```

```python
sample_rgb = [Image.open(f'{path}/rgb_{img}.png') for img in range(1314, 1318)]
sample_depth = [Image.open(f'{path}/depth_{img}.png') for img in range(1314, 1318)]
```

## Unprocessed single imgs

```python
fig, ax = plt.subplots(1,2, figsize = (21,10))
ax[0].imshow(np.asarray(sample_rgb[0], dtype=np.uint8))
ax[1].imshow(np.asarray(sample_depth[0], dtype=np.uint8))
plt.show()
```

## Edge detection filter

```python
fig, ax = plt.subplots(1,2, figsize = (21,10))
ax[0].imshow(np.asarray(sample_rgb[0].filter(ImageFilter.FIND_EDGES), dtype=np.uint8))
ax[1].imshow(np.asarray(sample_depth[0].filter(ImageFilter.FIND_EDGES), dtype=np.uint8))
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (21,10))
rgb_small = sample_rgb[0].filter(ImageFilter.FIND_EDGES)
rgb_small.thumbnail((60,80))
depth_small = sample_depth[0].filter(ImageFilter.FIND_EDGES)
depth_small.thumbnail((60,80))
ax[0].imshow(np.asarray(rgb_small, dtype=np.uint8))
ax[1].imshow(np.asarray(depth_small, dtype=np.uint8))
plt.show()
```

## Contour filter

```python
fig, ax = plt.subplots(1,2, figsize = (21,10))
ax[0].imshow(np.asarray(sample_rgb[0].filter(ImageFilter.CONTOUR), dtype=np.uint8))
ax[1].imshow(np.asarray(sample_depth[0].filter(ImageFilter.CONTOUR), dtype=np.uint8))
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (21,10))
rgb_small = sample_rgb[0].filter(ImageFilter.CONTOUR)
rgb_small.thumbnail((60,80))
depth_small = sample_depth[0].filter(ImageFilter.CONTOUR)
depth_small.thumbnail((60,80))
ax[0].imshow(np.asarray(rgb_small, dtype=np.uint8))
ax[1].imshow(np.asarray(depth_small, dtype=np.uint8))
plt.show()
```

## Enhance edge filter

```python
fig, ax = plt.subplots(1,2, figsize = (21,10))
ax[0].imshow(np.asarray(sample_rgb[0].filter(ImageFilter.EDGE_ENHANCE_MORE), dtype=np.uint8))
ax[1].imshow(np.asarray(sample_depth[0].filter(ImageFilter.EDGE_ENHANCE_MORE), dtype=np.uint8))
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (21,10))
rgb_small = sample_rgb[0].filter(ImageFilter.EDGE_ENHANCE_MORE)
rgb_small.thumbnail((60,80))
depth_small = sample_depth[0].filter(ImageFilter.EDGE_ENHANCE_MORE)
depth_small.thumbnail((60,80))
ax[0].imshow(np.asarray(rgb_small, dtype=np.uint8))
ax[1].imshow(np.asarray(depth_small, dtype=np.uint8))
plt.show()
```

## Stack imgs

```python
rgb_series = np.concatenate([np.asarray(sample, dtype=np.uint8) for sample in sample_rgb], axis=1)
depth_series = np.concatenate([np.asarray(sample, dtype=np.uint8) for sample in sample_depth], axis=1)

fig, ax = plt.subplots(2,1, figsize = (21,9))
ax[0].imshow(rgb_series)
ax[1].imshow(depth_series)
plt.show()
```

```python
rgb_series = np.concatenate([np.asarray(sample.filter(ImageFilter.EDGE_ENHANCE_MORE), dtype=np.uint8) for sample in sample_rgb], axis=1)
depth_series = np.concatenate([np.asarray(sample.filter(ImageFilter.EDGE_ENHANCE_MORE), dtype=np.uint8) for sample in sample_depth], axis=1)

fig, ax = plt.subplots(2,1, figsize = (21,9))
ax[0].imshow(rgb_series)
ax[1].imshow(depth_series)
plt.show()
```
